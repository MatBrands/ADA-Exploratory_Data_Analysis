```python
df_movies_clean = df_movies_raw[(df_movies_raw['Runtime'] != 'Unknown') & (df_movies_raw['Budget'] != 'Unknown') & (df_movies_raw['Income'] != 'Unknown')]
```

```python
df_movies_clean['Runtime_cat'] = np.nan
df_movies_clean.loc[(pd.to_numeric(df_movies_clean['Runtime'])>=0) & (pd.to_numeric(df_movies_clean['Runtime'])<60),'Runtime_cat'] = 'curto'
df_movies_clean.loc[(pd.to_numeric(df_movies_clean['Runtime'])>=60) & (pd.to_numeric(df_movies_clean['Runtime'])<120),'Runtime_cat'] = 'medio'
df_movies_clean.loc[(pd.to_numeric(df_movies_clean['Runtime'])>=120),'Runtime_cat'] = 'longo'
```

```python
regex_syntax = r"\D"

df_movies_clean['Budget'] = df_movies_clean['Budget'].str.replace('$', '')
```
